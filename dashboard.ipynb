```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


sns.set_style("darkgrid")
```

```python
df= pd.read_csv("sales_data . csv")
```

```python
df.head(5)
```

```python
sns.boxplot(x='Product', y='Price', data=df)
plt.title('Price Distribution by Category')
plt.show()
```

```python
sns.violinplot(x='Region', y='Total_Sales', data=df)
plt.title('Price Density by Category')
plt.show()
```

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

sns.boxplot(ax=axes[0,0], x='Product', y='Price', data=df)
sns.violinplot(ax=axes[0,1], x='Region', y='Total_Sales', data=df)
sns.histplot(ax=axes[1,0], x='Price', data=df)
sns.scatterplot(ax=axes[1,1], x='Price', y='Total_Sales', data=df)

plt.tight_layout()
plt.show()
```

```python
# Group data by region
region_sales = df.groupby('Region')['Total_Sales'].sum().reset_index()

# Plot
sns.barplot(x='Region', y='Total_Sales', data=region_sales)

plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Group by date
daily_sales = df.groupby('Date')['Total_Sales'].sum().reset_index()

# Plot
plt.figure(figsize=(10,5))
sns.lineplot(x='Date', y='Total_Sales', data=daily_sales)

plt.title('Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.show()
```
